```python
def dfedit (df_temp):
    """
    the functions returns a dataframe
    """
    cols = ["Nr.","Liefertermin","Umsatz","Stückpreis","Zahlziel"]
    df_temp.set_axis(cols,axis = 1,inplace=True)
    df_temp.drop(df_temp[df_temp["Nr."].isnull()].index,inplace=True)

    df_temp["Stück"] = df_temp.Umsatz.str.split(" ")
    df_temp["Umsatz"] = df_temp["Stück"].str[1].astype(int)
    df_temp["Stück"] = df_temp["Stück"].str[0].astype(int)
    df_temp["Zahlziel"] = df_temp["Zahlziel"].astype(int)
    df_temp["Nr."] = df_temp["Nr."].astype(int)
    df_temp.loc[df_temp["Liefertermin"].isnull(),'Liefertermin'] = "0"
    df_temp["Liefertermin"] = df_temp["Liefertermin"].str.replace('[\.\.]','').astype(int)
    df_temp["Stückpreis"] = df_temp["Stückpreis"].str.replace('[\,\.]','.').astype(float)

    return df_temp

```
